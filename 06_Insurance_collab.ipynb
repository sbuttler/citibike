## Number of cyclists injured

```python
print(f"Total number of cyclists injured: {sum(df.loc[:,'NUMBER OF CYCLIST INJURED'])}")
```

## Number of crashes that involve bikes with >1,000$ claims values

```python
df['bike_involved'] = df.loc[:, ['VEHICLE TYPE CODE 1','VEHICLE TYPE CODE 2','VEHICLE TYPE CODE 3','VEHICLE TYPE CODE 4','VEHICLE TYPE CODE 5']].isin(['Bike']).any(1)
```

```python
print(f"Total number of crashes (1,000$ damage): {df.loc[(df['NUMBER OF PERSONS KILLED']==0)&(df['NUMBER OF PERSONS INJURED']==0)&df['bike_involved'],:].shape[0]}")
```

```python
print(f"Percentage of crashes (1,000$ damage): {df.loc[(df['NUMBER OF PERSONS KILLED']==0)&(df['NUMBER OF PERSONS INJURED']==0)&df['bike_involved'],:].shape[0]/df.shape[0]*100:.2f}%")
```
